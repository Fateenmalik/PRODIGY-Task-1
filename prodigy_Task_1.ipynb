```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv(
    'API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv',
    skiprows=4
)
```

```python
print("Shape of Dataset:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nFirst 5 Rows:")
df.head()
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=['Unnamed: 69'])
```

```python
df.info()
```

```python
df.describe()
```

```python
population_2024 = df[['Country Name', '2024']]

population_2024.head()
```

```python
population_2024 = population_2024.dropna()

population_2024.shape
```

```python
top10 = population_2024.sort_values(
    by='2024',
    ascending=False
).head(10)

top10
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    top10['Country Name'],
    top10['2024']
)

plt.title(
    'Top 10 Countries by Population (2024)',
    fontsize=16
)

plt.xlabel('Country')
plt.ylabel('Population')

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(14,7))

sns.barplot(
    x='Country Name',
    y='2024',
    data=top10
)

plt.title(
    'Top 10 Most Populated Countries (2024)',
    fontsize=18
)

plt.xlabel('Country')
plt.ylabel('Population')

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.hist(
    population_2024['2024'],
    bins=20
)

plt.title(
    'Distribution of Country Populations',
    fontsize=16
)

plt.xlabel('Population')
plt.ylabel('Number of Countries')

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.hist(
    np.log10(population_2024['2024']),
    bins=20
)

plt.title(
    'Population Distribution (Log Scale)',
    fontsize=16
)

plt.xlabel('Log10 Population')
plt.ylabel('Number of Countries')

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.bar(
    top10['Country Name'],
    top10['2024']
)

plt.xticks(rotation=45)

plt.title(
    'Top 10 Countries by Population'
)

plt.tight_layout()

plt.savefig(
    'population_distribution.png',
    dpi=300
)

plt.show()
```
